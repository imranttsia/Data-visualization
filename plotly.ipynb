```python
import plotly.express as px
df1=px.data.tips()
display(df1.head())
fig=px.scatter(df1, x="total_bill", y="tip",color="smoker",facet_col="sex",facet_row="time")
fig.show()
```

```python
import plotly.express as px
df1=px.data.tips()
display(df1.head())
fig=px.scatter(df1, x="total_bill", y="tip",color="smoker",facet_col="sex",facet_row="time",text="tip")
fig.show()
```
