```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x1 = np.sort(np.random.choice(x,20, replace=False))
y1 = np.sort(np.random.choice(x,20, replace=False))

display(x1)
display(y1)

plt.plot(x1,y1)
plt.show()
```

```python
pak_score = np.random.randint(1,36, 20)
ind_score = np.random.randint(1,36, 20)

overs = np.arange(1,21)

display(pak_score)
display(ind_score)


plt.plot(overs, pak_score, label='Pakistan')
plt.plot(overs, ind_score, label="India")
plt.title("Pakistan Vs India")
plt.xlabel("Overs")
plt.ylabel("Score")
plt.legend()
plt.show()
```
